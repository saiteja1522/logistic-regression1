```python
# prompt: convert the string values of column 'Name' to numerical values

import pandas as pd

# Assuming your DataFrame is named 'data'
# Create a mapping of unique names to numerical values
unique_names = data['Name'].unique()
name_to_numerical = {name: i for i, name in enumerate(unique_names)}

# Replace the string values in the 'Name' column with numerical values
data['Name'] = data['Name'].map(name_to_numerical)

# Print the updated DataFrame
print(data.head())
```

```python
# prompt: covert all column values to numerical values

# Assuming your DataFrame is named 'data'

for column in data.columns:
  if data[column].dtype == object:
    # Create a mapping of unique string values to numerical values
    unique_values = data[column].unique()
    value_to_numerical = {value: i for i, value in enumerate(unique_values)}

    # Replace the string values in the column with numerical values
    data[column] = data[column].map(value_to_numerical)

# Print the updated DataFrame
print(data.head())
```

```python
# Create a correlation matrix heatmap to visualize the correlations between numerical features.
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Build a logistic regression model using appropriate libraries (e.g., scikit-learn).

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Assuming 'data' is your DataFrame with preprocessed data
# Define your features (X) and target variable (y)
X = data.drop('Survived', axis=1)  # Remove the 'Survived' column to get features
y = data['Survived']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate a classification report
print("Classification Report:\n", classification_report(y_test, y_pred))

# Generate a confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
#Evaluate the performance of the model on the testing data using accuracy, precision, recall, F1-score, and ROC-AUC score.

from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score


# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate Precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Calculate Recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# Calculate F1-Score
f1 = f1_score(y_test, y_pred)
print("F1-Score:", f1)

# Calculate ROC-AUC score
try:
  y_pred_proba = model.predict_proba(X_test)[:, 1]  # Get probability of positive class
  roc_auc = roc_auc_score(y_test, y_pred_proba)
  print("ROC-AUC Score:", roc_auc)
except AttributeError:
  print("ROC-AUC score cannot be calculated for this model.")
```

```python

from sklearn.metrics import roc_curve, auc


fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
